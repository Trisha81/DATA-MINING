```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
#step1 :load the iris dataset
iris = load_iris()
X= iris.data
y= iris.target
```

```python
#step2: preprocess the data (standardization)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#step3: apply pca to reduce to 2 dimensions
pca= PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
#step4: visualize the reduced features (2d sactter plot)
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap=plt.cm.get_cmap('viridis'), edgecolors='k',s=100)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Visualization of Iris Dataset')
plt.colorbar(label='Tareget Class')
plt.show()
```

```python
#optional : Explained varience ratio
print("Explained varience ratio of each component: ")
print(pca.explained_variance_ratio_)
```
